```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\next3\Downloads\shopping_trends_updated.csv")
df.head()
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
sales_per_location = df.groupby("Location")["Purchase Amount (USD)"].sum().reset_index()

sales_per_location = sales_per_location.sort_values(by="Purchase Amount (USD)", ascending=False)
sales_per_location
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Location", y="Purchase Amount (USD)", data=sales_per_location, palette="viridis")
plt.title("Total Sales per Location")
plt.xlabel("Location")
plt.ylabel("Total Sales (USD)")
plt.xticks(rotation=45)
plt.show()
```

```python

sales_gender = df.groupby("Gender")["Purchase Amount (USD)"].sum()
plt.figure(figsize=(6,6))
plt.pie(sales_gender, labels=sales_gender.index, autopct='%1.1f%%', colors=['skyblue', 'lightpink'], startangle=90)
plt.title("Sales Distribution by Gender")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x="Item Purchased", hue="Gender", data=df)
plt.title("Number of Purchases per Product by Gender")
plt.xlabel("Product")
plt.ylabel("Count")
plt.xticks(rotation=45)  
plt.legend(title="Gender")
plt.show()
```

```python
sales_per_season = df.groupby("Season")["Purchase Amount (USD)"].sum().reset_index()
plt.figure(figsize=(8,5))
sns.barplot(x="Season", y="Purchase Amount (USD)", data=sales_per_season, palette="coolwarm")
plt.title("Total Sales per Season")
plt.xlabel("Season")
plt.ylabel("Total Sales (USD)")
plt.show()
```

```python
sales_per_size = df.groupby("Size")["Purchase Amount (USD)"].sum().reset_index()

plt.figure(figsize=(8,5))
sns.barplot(x="Size", y="Purchase Amount (USD)", data=sales_per_size, palette="Set2")
plt.title("Total Sales by Size")
plt.xlabel("Size")
plt.ylabel("Total Sales (USD)")
plt.show()
```

```python
sales_per_shipping = df.groupby("Shipping Type")["Purchase Amount (USD)"].sum().reset_index()

plt.figure(figsize=(8,5))
sns.barplot(x="Shipping Type", y="Purchase Amount (USD)", data=sales_per_shipping, palette="Set2")
plt.title("Total Sales by Shipping Type")
plt.xlabel("Shipping Type")
plt.ylabel("Total Sales (USD)")
plt.show()
```

```python
sales_per_color = df.groupby("Color")["Purchase Amount (USD)"].sum().reset_index()

plt.figure(figsize=(10,6))
sns.barplot(x="Color", y="Purchase Amount (USD)", data=sales_per_color, palette="tab20")
plt.title("Total Sales by Color")
plt.xlabel("Color")
plt.ylabel("Total Sales (USD)")
plt.xticks(rotation=45)
plt.show()
```

```python
age_bins = [0, 20, 30, 40, 50, 60, 70]
df['Age Group'] = pd.cut(df['Age'], bins=age_bins)

sales_per_age_group = df.groupby("Age Group")["Purchase Amount (USD)"].sum().reset_index()

plt.figure(figsize=(8,5))
sns.barplot(x="Age Group", y="Purchase Amount (USD)", data=sales_per_age_group, palette="Set2")
plt.title("Total Sales by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Total Sales (USD)")
plt.show()
```

```python
sales_per_frequency = df.groupby("Frequency of Purchases")["Purchase Amount (USD)"].sum().reset_index()
plt.figure(figsize=(8,5))
sns.barplot(x="Frequency of Purchases", y="Purchase Amount (USD)", data=sales_per_frequency, palette="Set2")
plt.title("Total Sales by Purchase Frequency")
plt.xlabel("Purchase Frequency")
plt.ylabel("Total Sales (USD)")
plt.show()
```

```python
sales_per_promo = df.groupby("Promo Code Used")["Purchase Amount (USD)"].sum().reset_index()
plt.figure(figsize=(6,5))
sns.barplot(x="Promo Code Used", y="Purchase Amount (USD)", data=sales_per_promo, palette="Set3")
plt.title("Total Sales by Promo Code Usage")
plt.xlabel("Promo Code Used")
plt.ylabel("Total Sales (USD)")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x="Season", y="Purchase Amount (USD)", hue="Category", data=df, palette="Set2")
plt.title("Total Sales per Category in Each Season")
plt.xlabel("Season")
plt.ylabel("Total Sales (USD)")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["Purchase Amount (USD)"], bins=20, kde=True, color="skyblue")
plt.axvline(df["Purchase Amount (USD)"].median(), color="red", linestyle="--", label=f"Median = {df['Purchase Amount (USD)'].median():.2f}")
plt.title("Distribution of Purchase Amount (USD)")
plt.xlabel("Purchase Amount (USD)")
plt.ylabel("Number of Customers")
plt.legend()
plt.show()
```
